```python
from sklearn.preprocessing  import PolynomialFeatures
poly=PolynomialFeatures(degree=2,include_bias=False)
x_train_poly = poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
poly_feature_names = poly.get_feature_names_out(x_train.columns)
y_train_flat=y_train.values.ravel()
y_test_flat=y_test.values.ravel()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
param_dist={'alpha':[0.1,1.0,10.0,100.0]}
ridge=Ridge()
Grid_ridge=GridSearchCV(ridge,param_dist,cv=5)
Grid_ridge.fit(x_train_poly,y_train_flat)
```

```python
Grid_ridge.score(x_test_poly,y_test_flat)
```

```python
y_pred_ridge=Grid_ridge.predict(x_test_poly)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, accuracy_score, precision_recall_fscore_support
import numpy as np
mse = mean_squared_error(y_test, y_pred_ridge)
mae = mean_absolute_error(y_test, y_pred_ridge)
print(f"MSE: {mse}")
print(f"MAE: {mae}")
r2 = r2_score(y_test, y_pred_ridge)
print(f"R² Score: {r2}")
```
